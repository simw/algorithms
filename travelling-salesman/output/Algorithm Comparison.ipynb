```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Comparing all algorithms
---------

2= Nearest neighbour
3= Insertion heuristic
4= Edge swap
5= Probabilistic distance weighted
6= Ant colony

```python
data = pd.read_csv('testall.txt')
lux = pd.read_csv('testall_luxembourg_980.txt')
```

```python
fig1, axs1 = plt.subplots(1,2, figsize=(18,8))

data.plot(x='algorithm',y='distance',kind='scatter',ax=axs1[0])
axs1[0].set_title('sample.tsp')

lux.plot(x='algorithm',y='distance',kind='scatter',ax=axs1[1])
axs1[1].set_title('luxembourg_980.tsp')
```

```python
fig2, axs2 = plt.subplots(1, 2, figsize=(18,8))

data[data['algorithm']!=1].boxplot(by='algorithm',column='distance',ax=axs2[0])
axs2[0].set_title('sample.txt')
for i in [2,3,4,5,6]:
    y = data.distance[data.algorithm==i].dropna()
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i-1, 0.04, size=len(y))
    axs2[0].plot(x, y, 'r.', alpha=0.5,markersize=20)
    
lux[lux['algorithm']!=1].boxplot(by='algorithm',column='distance',ax=axs2[1])
axs2[1].set_title('luxembourg_980.tsp')
for i in [2,3,4,5,6]:
    y = lux.distance[lux.algorithm==i].dropna()
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i-1, 0.04, size=len(y))
    axs2[1].plot(x, y, 'r.', alpha=0.5,markersize=20)
```

Comparing distance weighting exponents (algo 5)

```python
distance_data.plot(x='exponent',y='distance',kind='scatter',figsize=(12,8))
```

```python
distance_data[distance_data.exponent > offset].exponent.unique()
```

```python
offset = 2
distance_data[distance_data.exponent > offset].boxplot(by='exponent',column='distance',figsize=(12,8))

def xaxis(x):
    if x < 11:
        return x
    else:
        return 11 + (x-12)/4
    
for i in distance_data[distance_data.exponent > offset].exponent.unique():
    y = distance_data.distance[distance_data.exponent==i]
    # Add some random "jitter" to the x-axis
    x = np.random.normal(xaxis(i)-offset, 0.04, size=len(y))
    plt.plot(x, y, 'r.', alpha=0.5,markersize=20)
```

Ants vs edges
------

```python
ants = pd.read_csv('testants.txt')
edges = pd.read_csv('testrandom.txt')
```

```python
ants['label'] = ants['n'].map('n={0:04d} ants'.format)
edges['label'] = edges['n'].map('n={0:04d} edges'.format)

alldata = pd.concat([ants, edges])
alldata = alldata.loc[:, ['label', 'distance']]
```

```python
alldata.boxplot(by='label', column='distance',figsize=(12,8))
i = 0
for label in alldata.sort('label').label.unique():
    i = i + 1
    y = alldata.distance[alldata.label==label]
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i, 0.04, size=len(y))
    plt.plot(x, y, 'r.', alpha=0.5,markersize=20)
```

```python
with open('testedges.txt') as f:
    line_number = 0
    dists = []
    courses = []
    for line in f:
        line_number += 1
        if (line_number - 1) % 4 == 2:
            dists.append(float(line.strip('Distance =').strip()))
        elif (line_number - 1) % 4 == 3:
            courses.append(line)
            
print(len(dists))
print(min(dists))
print(courses[dists.index(min(dists))])
```
